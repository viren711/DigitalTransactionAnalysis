```python
import pandas as pd

csv_files = ['18.csv', '19.csv', '20.csv', '21.csv', '22.csv', '23.csv', '24.csv', '25.csv']
years = ['2018-2019', '2019-2020', '2020-2021', '2021-2022', '2022-2023', '2023-2024', '2024-2025', '2025-2026']

dfs = []
for fname, year in zip(csv_files, years):
    df = pd.read_csv(fname, index_col=0)
    df.columns = [year]
    dfs.append(df)

# Outer join to preserve all Modes, fill missing data with 0
combined = pd.concat(dfs, axis=1, join='outer').fillna(0)

# Reset index so that Mode is a column, not index
combined.reset_index(inplace=True)

# Save to CSV
combined.to_csv('digital_payments.csv', index=False)

print('Combined CSV with zeros for missing entries saved as all_years_horizontal_matched_zero.csv')
```
