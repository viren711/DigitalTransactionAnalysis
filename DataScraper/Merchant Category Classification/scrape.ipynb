```python
import requests
import pandas as pd

url = "https://www.npci.org.in/api/ecosystem-statistics/get-statistics?product_name=UPI&tab_name=chargeback&year=2025&month=Mar&page_no=1&sort_by=asc&size=1000&locale=en"
url = "https://www.npci.org.in/api/ecosystem-statistics/get-statistics?product_name=UPI&tab_name=chargeback&year=2025&month=Jan&page_no=1&sort_by=asc&size=10&locale=en"
resp = requests.get(url)
data = resp.json()

# Check how your data is nested.
if 'data' in data and isinstance(data['data'], list):
    records = data['data']
elif isinstance(data, list):
    records = data
else:
    print("Couldn't find list data, please inspect API response structure.")
    records = []

# Normalize to ensure all dicts have the same keys
import collections

# Find all keys that appear anywhere
all_keys = set()
for record in records:
    if isinstance(record, dict):
        all_keys.update(record.keys())

# Fill missing keys for each record
normalized_records = []
for record in records:
    if isinstance(record, dict):
        normalized_records.append({k: record.get(k, None) for k in all_keys})

# Now create the DataFrame
df = pd.DataFrame(normalized_records)

# Save to CSV
df.to_csv("upi_chargeback_stats_fixed.csv", index=False)
print("Saved upi_chargeback_stats_fixed.csv")
```
