```python
import requests

# Example: Replace this with the actual link you found (Excel/CSV download)
download_url = "https://www.npci.org.in/PATH/TO/DATA.xlsx"
file_name = "downloaded_mcc_data.xlsx"

response = requests.get(download_url)
with open(file_name, "wb") as f:
    f.write(response.content)

print("File downloaded:", file_name)
```
